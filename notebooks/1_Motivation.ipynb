<h1>Data vs. theory in psychology</h1>
<li class="fragment level1">Observation: Humans have a very poor understanding of randomness</li>
	<li class="fragment level2">The sequence HHHHHHHH is judged as less probable than HTHTTHTH</li>
<li class="fragment level1">Explanations for this observation:</li>
	<li class="fragment level2">The representativeness heuristic: we expect small samples to accurately represent the properties of the underlying population</li>
	<li class="fragment level2">Maybe HHHHHHHH *is* actually less probable than HTHTTHTH--the former is consistent with cheating!</li>
	<li class="fragment level2">When the aliens built us, they implemented very bad pseudorandom number generators</li>
<li class="fragment level1">We call these things 'data' and 'theories', respectively</li>


<h1>Theory gets most of the credit</h1>
<blockquote class="fragment bq">Theory is widely regarded as the most creative form of scientific contribution. Scientific disciplines that stress theory are characterized as "basic" or "pure." More empirically or practically focused disciplines are seen as "technical" or "applied"--labels that most will see as implying lower status. Journals that feature theoretical articles often stand as their disciplines' publication flagships. Psychology's most elite empirical journals often oblige authors to establish the value of submitted articles by making clear how their empirical work "advances theory." ([Greenwald, 2012](http://faculty.washington.edu/agg/pdf/NothingSoTheoretical.PPS.2012.pdf), p.99)</blockquote>

<h1>In practice...</h1>
<li class="fragment level1">Theoretical debates are rarely resolved; typically die out gradually</li>

<h1>A (seemingly) simple recipe</h1>
<li class="fragment level1">Generate a hypothesis</li>
<li class="fragment level1">Design an experiment</li>
<li class="fragment level1">Collect some data</li>
<li class="fragment level1">Analyze data, find interesting result(s) (p < .05)</li>
<li class="fragment level1">Publish result(s)</li>
<li class="fragment level1">Repeat</li>
<br /><li class="fragment level1">Problems arise at every stage of this process</li>


<h1>Data collection</h1>

<h1>Managing and organizing data</h1>
<li class="fragment level1">Most scientists don't manage their data/analyses/results very systematically</li>
	<li class="fragment level2">Who did what analysis? When? How were research decisions made? Why are there four versions of the data?</li>
<li class="fragment level1">Most experimental designs and findings are not annotated in a machine-readable way</li>
	<li class="fragment level2">How do we find all studies done in population X? What's the relationship between concepts A and B?</li>
<li class="fragment level1">Compare with, e.g., genetics</li>


<h1>Data organization, standardization, and annotation</h1>
<li class="fragment level1">Encourage version control for everything (GitHub, OSF, etc.)</li>
<li class="fragment level1">Promote use of automated data processing and analysis pipelines</li>
	<li class="fragment level2">[NiPype](http://nipy.sourceforge.net/nipype/), [scikit-learn](http://scikit-learn.org)</li>
<p class="fragment level1"><img src="images/cognitive_atlas.jpg" class='float-right' style='width: 35%;'></p>
<li class="fragment level1">Develop formal ontologies of psychological tasks, processes, etc.</li>
	<li class="fragment level2">E.g., the [Cognitive Atlas](http://cognitiveatlas.org)</li>


<h1>Bigger, better analyses</h1>
<p class="fragment level1"><img src="images/ioannidis.png" class='float-right'></p>
<li class="fragment level1">Conventional statistical analyses in psychology rely heavily on hypothesis testing</li>
	<li class="fragment level2">Little attention to effect size</li>
	<li class="fragment level2">Overfitting is very common ([Ioannidis, 2008](http://dcscience.net/ioannidis-associations-2008.pdf))</li>
<li class="fragment level1">Other approaches can play a complementary role</li>
	<li class="fragment level2">Acquisition of larger, richer datasets</li>
	<li class="fragment level2">Greater emphasis on cross-validation and out-of-sample prediction</li>
	<li class="fragment level2">Machine learning techniques that are less susceptible to multicollinearity</li>
	<li class="fragment level2">Computationally-intensive hierarchical statistical models</li>


<h1>Faster, more effective evaluation</h1>
<li class="fragment level1">Publishing is a frustratingly slow process</li>
<li class="fragment level1">The typical review process in psychology takes > 6 months</li>
<li class="fragment level1">Reliability of conventional peer review is low ([Bornmann et al, 2010](http://www.plosone.org/article/info%3Adoi%2F10.1371%2Fjournal.pone.0014331))</li>
<li class="fragment level1">New publishing and evaluation models are rapidly emerging</li>
	<li class="fragment level2">E.g., [PeerJ](https://peerj.com/), [PubMed Commons](http://www.ncbi.nlm.nih.gov/pubmedcommons/), [PubPeer](http://pubpeer.com), [hypothes.is](http://hypothes.is/), etc.</li>

<h1>The actor-observer bias</h1>
<li class="fragment level1">Jones & Nisbett (1971): "actors tend to attribute the causes of their behavior to stimuli inherent in the situation, while observers tend to attribute behavior to stable dispositions of the actor"</li>
<li class="fragment level1">Extremely influential, but is it true?</li>


<h1>On fooling ourselves</h1>
<blockquote class="fragment bq"><p>We have learned a lot from experience about how to handle some of the ways we fool ourselves. One example: Millikan measured the charge on an electron by an experiment with falling oil drops, and got an answer which we now know not to be quite right. It's a little bit off, because he had the incorrect value for the viscosity of air. It's interesting to look at the history of measurements of the charge of the electron, after Millikan. If you plot them as a function of time, you find that one is a little bigger than Millikan's, and the next one's a little bit bigger than that, and the next one's a little bit bigger than that, until finally they settle down to a number which is higher.</p><p>Why didn't they discover that the new number was higher right away? It's a thing that scientists are ashamed of--this history--because it's apparent that people did things like this: When they got a number that was too high above Millikan's, they thought something must be wrong--and they would look for and find a reason why something might be wrong. When they got a number closer to Millikan's value they didn't look so hard. And so they eliminated the numbers that were too far off, and did other things like that. We've learned those tricks nowadays, and now we don't have that kind of a disease.<br />--Richard Feynman, [Cargo Cult Science](http://neurotheory.columbia.edu/~ken/cargo_cult.html)</p></blockquote>

<h1>The actor-observer bias</h1>
<blockquote class="fragment bq">Widely known in psychology, this asymmetry has been described as robust, firmly established, and pervasive. However, a meta-analysis on 173 published studies revealed average effect sizes from d = -0.016 to d = 0.095. A moderator analysis showed that the asymmetry held only when the actor was portrayed as highly idiosyncratic, when hypothetical events were explained, when actor and observer were intimates, or when free-response explanations were coded. In addition, the asymmetry held for negative events, but a reverse asymmetry held for positive events. This valence effect may indicate a self-serving pattern in attribution, but across valence, no actor-observer asymmetry exists. ([Malle, 2006](http://research.clps.brown.edu/soccogsci/Publications/Pubs/Malle_%282006%29_ActObs_meta.pdf))</blockquote><li class="fragment level1">One meta-analysis is no match for 40 years of received wisdom</li>
<li class="fragment level1">How long should we wait for the truth to emerge?</li>

<h1>A working definition</h1>
<p class="fragment level1"><em>Psychoinformatics, n.</em> The use of modern computing technologies and methods to improve the acquisition, organization, synthesis, dissemination, and evaluation of psychological data.</p>


<h1>Relationship to...</h1>
<li class="fragment level1">Quantitative psychology?</li>
	<li class="fragment level2">Quantitative psychology focuses on measurement/psychometrics</li>
<li class="fragment level1">Computational psychology?</li>
	<li class="fragment level2">Computational psychology focuses on theoretical modeling of cognitive processes</li>
<li class="fragment level1">Data science?</li>
	<li class="fragment level2">Data science focuses on processing/analysis of data, not platform-building</li>
